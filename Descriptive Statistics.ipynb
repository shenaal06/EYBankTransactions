```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load and preprocess data
data2 = pd.read_csv("C:/Users/User/Bank Transactions Analysis/BankTransactionData_Cleaned.csv")

# Display the first few rows of the dataset
print("Dataset Preview:\n", data2.head())
```

```python
# 1. Histogram for Transaction Amount
plt.figure(figsize=(10, 6))
sns.histplot(data2['TransactionAmount'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Boxplot for Customer Age
plt.figure(figsize=(10, 6))
sns.boxplot(x=data2['CustomerAge'], color='lightgreen')
plt.title('Boxplot of Customer Age')
plt.xlabel('Customer Age')
plt.show()
```

```python
# 3. Scatter Plot for Transaction Amount vs Account Balance
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data2['TransactionAmount'], y=data2['AccountBalance'], color='orange')
plt.title('Transaction Amount vs Account Balance')
plt.xlabel('Transaction Amount')
plt.ylabel('Account Balance')
plt.show()
```

```python
# 4. Heatmap of Correlation Matrix
plt.figure(figsize=(8, 6))
correlation_matrix = data2.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
